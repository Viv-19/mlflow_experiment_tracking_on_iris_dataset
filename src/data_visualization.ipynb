```python
import matplotlib.pyplot as plt
import seaborn as sns
import os
import pandas as pd
```

```python
# Step 2: Load pre-saved CSV from data_ingestion
data_path = "../data/iris_data.csv"
if os.path.exists(data_path):
    df = pd.read_csv(data_path)
else:
    raise FileNotFoundError("Data file not found. Run data_ingestion.ipynb first.")
```

```python
# Step 3: Countplot - Sample distribution per class
plt.figure(figsize=(4, 2))
sns.countplot(x='target_name', data=df, palette='Set1')
plt.title('Count of Each Iris Species')
plt.xlabel('Species')
plt.ylabel('Count')
plt.show()
```

```python
# Step 4: Pairplot - Feature relationships (excluding 'target' column)
sns.pairplot(df.drop(columns=['target']), hue='target_name', palette='Set1', height=2)
plt.suptitle("Pairplot of Iris Features", y=1.02)
plt.show()
```

```python
# Step 5: Correlation heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(df.iloc[:, :-2].corr(), annot=True, cmap='coolwarm')
plt.title("Feature Correlation Heatmap")
plt.show()
```
